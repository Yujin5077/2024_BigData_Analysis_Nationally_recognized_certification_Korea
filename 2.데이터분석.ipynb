### 2. 데이터 분석
#### 2.1. 분류
  - 빅분기 실기에서의 데이터 분석 파트는 랜덤 포레스트를 활용한 풀이가 정확도가 높으면서 점수가 잘나오는 정설로 자리잡혔음
  - 이러한 이유로 분류와 회귀 문제에서의 랜덤 포레스트 모델을 사용하는 방법을 정리할 것임

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

#랜덤포레스트 분류 - 타이타닉 데이터셋
from sklearn.ensemble import RandomForestClassifier

# 학습 및 테스트 데이터 분리
from sklearn.model_selection import train_test_split

#데이터 셋
df = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사_실기/train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# 결측치 평균값 대치
d_mean= df["Age"].mean()
df["Age"].fillna(d_mean, inplace=True)

# 결측치 최빈값 대치
d_mode = df["Embarked"].mode()[0]
df["Embarked"].fillna(d_mode, inplace=True)
```

```python
# 레이블 인코딩
# Sex 컬럼과 Embarked 컬럼을 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

df["Sex"] = LabelEncoder().fit_transform(df["Sex"])
df["Embarked"] = LabelEncoder().fit_transform(df["Embarked"])
```

```python
# SibSp, Parch의 값을 더해서 FamilySize 파생변수를 생성
df["FamilySize"] = df["SibSp"] + df["Parch"]
```

```python
# 분석 데이터셋 준비
# x는 독립변수, y는 종속변수
x = df[["Pclass", "Sex", "Age", "Fare", "Embarked", "FamilySize"]]
y = df["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=15)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=3, random_state=20)
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

#### 2.2. 회귀
- 랜덤포레스트 회귀 모델을 사용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사_실기/Housing.csv")
df.head()
```

```python
df.info()
```

```python
df = df.dropna(axis=0)

df = df.drop("mainroad", axis=1)
df = df.drop("guestroom", axis=1)
df = df.drop("basement", axis=1)
df = df.drop("hotwaterheating", axis=1)
df = df.drop("airconditioning", axis=1)
df = df.drop("prefarea", axis=1)
df = df.drop("furnishingstatus", axis=1)
```

```python
corr = df.corr(method="pearson")
print(corr)
```

```python
x = df.drop("price", axis=1)
y = df["price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf = RandomForestRegressor(max_depth=3, random_state=15)
rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

# 모델 성능평가 - 테스트 데이터셋
pred = rf.predict(x_test)
mse = mean_squared_error(y_test, pred)
print(mse)

# 모델 성능평가 - 학습 데이터셋
pred = rf.predict(x_train)
mse = mean_squared_error(y_train, pred)
print(mse)
```
